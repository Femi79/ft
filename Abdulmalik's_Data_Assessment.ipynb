```python
#Changing the datatype of Date
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Separating Months and Days from Date
df['Months'] = df['Date'].dt.month
df['Days'] = df['Date'].dt.day
```

```python
 # Define a function to get season based on month
def season(months):
        if months in [12, 1, 2]:
            return 'Winter'
        elif months in [3, 4, 5]:
            return 'Spring'
        elif months in [6, 7, 8]:
            return 'Summer'
        else:
            return 'Fall'
df['Season'] = df['Months'].apply(season)
```

**QUESTION 3**

```python
subset_data = df[['Months', 'Days', 'Season', 'Description']]
crime_counts = subset_data.groupby(['Months', 'Days', 'Season', 'Description']).size().reset_index(name='Frequency')
crime_counts.head()
```

```python
#determining the months, days and season with the highest crime rate
most_crime_month = crime_counts.groupby('Months')['Frequency'].sum().idxmax()
most_crime_day = crime_counts.groupby('Days')['Frequency'].sum().idxmax()
most_crime_season = crime_counts.groupby('Season')['Frequency'].sum().idxmax()

print("Month with the most crime records:", most_crime_month)
print("Day with the most crime records:", most_crime_day)
print("Season with the most crime records:", most_crime_season)
```

```python
#determining the most common crime and least common crime
crime_freq_summary = crime_counts.groupby('Description')['Frequency'].sum().reset_index()
most_common_crime = crime_freq_summary.loc[crime_freq_summary['Frequency'].idxmax()]
least_common_crime = crime_freq_summary.loc[crime_freq_summary['Frequency'].idxmin()]

print("Most common crime:", most_common_crime['Description'], "with", most_common_crime['Frequency'], "occurrences")
print("Least common crime:", least_common_crime['Description'], "with", least_common_crime['Frequency'], "occurrences")
```
